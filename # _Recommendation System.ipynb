```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling as pp
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
book = pd.read_csv("books.csv")
```

```python
book
```

```python
# taking columns from 1
book1 = book.iloc[:,1:]
```

```python
book1
```

```python
book2 = book1.rename({'User.ID':'User_ID','Book.Title':'Book_Title','Book.Rating':'Book_Rating'}, axis = 1)
```

```python
book2
```

```python
#number of unique users in the dataset
len(book2.User_ID.unique())
```

```python
# Knowing the length of the dataset
len(book2.Book_Title.unique())
```

```python
# Value count of movies length
book2.Book_Title.value_counts()
```

```python
#convrting data from long format to wide format using pivot_table
book3 = book2.pivot_table(index='User_ID',
                            columns='Book_Title',
                            values='Book_Rating').reset_index(drop=True)
```

```python
book3
```

```python
#Impute those NaNs with 0 values
book3.fillna(0, inplace=True)
```

```python
book3
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
# sim is none but similarities
user_sim = 1 - pairwise_distances(book3.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#assigning unique userid to rows and columns
user_sim_df.index = book2['User_ID'].unique()
user_sim_df.columns = book2['User_ID'].unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
#filling digonal values
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#finding users with max similarities
user_sim_df.idxmax(axis=1)
```
